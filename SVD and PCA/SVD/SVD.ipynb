# SVD

```python
import numpy as np
from matplotlib.image import imread
import matplotlib.pyplot as plt
import os
```

```python
A=imread('istockphoto-1256494945-612x612.jpg')
plt.imsave('grey_scale_husky.png',A,cmap='grey')
```

```python
X=np.mean(A,-1)
img=plt.imshow(256-X)
img.set_cmap('gray')
plt.axis('off')
plt.show()
```

```python
# Perform SVD
U, S, VT = np.linalg.svd(X, full_matrices=False)  # Economy SVD
S = np.diag(S)

fig, ax = plt.subplots(nrows=1, ncols=10, figsize=(15, 20))
l = np.arange(10, 210, 20)
for i, r in enumerate(l):
    # Calculate approximation
    Xapprox = U[:, :r] @ S[0:r, :r] @ VT[:r, :]

    # Plot the image in the corresponding subplot
    img = ax[i].imshow(256 - Xapprox)
    img.set_cmap('gray')
    ax[i].axis('off')
    ax[i].set_title('r=' + str(r))

plt.tight_layout()
plt.show()
```

```python
plt.figure(1)
plt.semilogy(np.diag(S))
plt.title('Singular Value')
plt.show()

plt.figure(2)
plt.plot(np.cumsum(np.diag(S))/np.sum(np.diag(S)))
plt.title('Singular Value: Cumulative Sum')
plt.show()
#At rank r, how much of the matrices is captured by the first r nodes
```
